```python
import os
from selenium import webdriver
from selenium.webdriver.common.by import By
from time import sleep
from selenium.common.exceptions import NoSuchElementException

def extract_data_from_links(links):
    path = "c://chromedriver.exe"
    
    # Open the browser
    browser = webdriver.Chrome()

    titles_folder = "titles"
    articles_folder = "articles"

    os.makedirs(titles_folder, exist_ok=True)
    os.makedirs(articles_folder, exist_ok=True)
    
    for idx, link in enumerate(links, start=1):
        try:
            # Load the webpage
            browser.get(link)
            browser.maximize_window()

            # Extract title
            title = browser.find_element(By.CLASS_NAME, "td-post-title").find_element(By.CLASS_NAME, ["entry-title", "tdb-title-text"]).text

            # Extract article content
            article_element = browser.find_element(By.CSS_SELECTOR, 'div[class="td-post-content tagdiv-type"]')
            article = article_element.text

            # Find the index of the starting word "Blackcoffer" in the article content
            start_index = article.find("Blackcoffer")

            # Remove everything starting from the found index
            cleaned_article = article[:start_index]

            # Store the data in separate folders and files
            title_filename = os.path.join(titles_folder, f"Title_{idx}.txt")
            article_filename = os.path.join(articles_folder, f"Article_{idx}.txt")

            with open(title_filename, 'w', encoding='utf-8') as title_file:
                title_file.write(title)

            with open(article_filename, 'w', encoding='utf-8') as article_file:
                article_file.write(cleaned_article)

            print(f"Data for link {idx} stored successfully.")

        except NoSuchElementException as e:
            # Handle the exception if the required elements are not found on the webpage
            print(f"Error processing {link}: {e}")

            try:
                # Retry extraction in the finally block
                # Extract title
                title = browser.find_element(By.CLASS_NAME, ["tdb-title-text"]).text

                # Extract article content
                article_element = browser.find_element(By.CSS_SELECTOR, 'div[class="td_block_wrap tdb_single_content tdi_130 td-pb-border-top td_block_template_1 td-post-content tagdiv-type"]')
                article = article_element.text

                # Find the index of the starting word "Blackcoffer" in the article content
                start_index = article.find("Blackcoffer")

                # Remove everything starting from the found index
                cleaned_article = article[:start_index]

                # Store the data in separate folders and files
                title_filename = os.path.join(titles_folder, f"Title_{idx}.txt")
                article_filename = os.path.join(articles_folder, f"Article_{idx}.txt")

                with open(title_filename, 'w', encoding='utf-8') as title_file:
                    title_file.write(title)

                with open(article_filename, 'w', encoding='utf-8') as article_file:
                    article_file.write(cleaned_article)

                print(f"Data for link {idx} (in finally block) stored successfully.")

            except NoSuchElementException as e:
                # Handle the exception if the required elements are not found on the webpage in finally block
                print(f"Error processing {link} in finally block: {e}")

            except Exception as e:
                # Handle other exceptions in finally block
                print(f"An unexpected error occurred in finally block: {e}")

    # Close the browser after processing all links
    browser.quit()

# Call the function with your list of links
Links = ["https://insights.blackcoffer.com/rising-it-cities-and-its-impact-on-the-economy-environment-infrastructure-and-city-life-by-the-year-2040-2/",
        "https://insights.blackcoffer.com/rising-it-cities-and-their-impact-on-the-economy-environment-infrastructure-and-city-life-in-future/",
        # ... (other links)
        "https://insights.blackcoffer.com/how-will-covid-19-affect-the-world-of-work-2/"]

extract_data_from_links(Links)
```
